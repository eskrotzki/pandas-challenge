```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Values 
total_schools = len(school_data_complete['school_name'].unique())

total_students = school_data_complete['student_name'].count()

total_budget = school_data_complete['budget'].sum()

average_math = round(school_data_complete['math_score'].mean(),3)

average_read = round(school_data_complete['reading_score'].mean(),3)

pass_math = school_data_complete.loc[school_data_complete['math_score'] >= 70,:]
percent_pass_math = round((len(pass_math) / total_students) * 100,3)

pass_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70,:]
percent_pass_reading = round((len(pass_reading) / total_students) * 100,3)

overall_passing = (percent_pass_math + percent_pass_reading) / 2
```

```python
summary_table_df = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], "Total Budget": [total_budget],
                              "Average Math Score": [average_math], "Average Reading Score": [average_read], "Percent Passing Math": [percent_pass_math],
                              "Percent Passing Reading": [percent_pass_reading], "Percent Overall Passing Rate": [overall_passing]})
```

```python
summary_table_df["Total Budget"] = summary_table_df["Total Budget"].map("${:,.2f}".format)
summary_table_df
```

```python

group_school = school_data_complete.groupby(["school_name"])

# Values
school_name1 = school_data.set_index(["school_name"])["type"]

total_students1 = group_school["student_name"].count()

total_budget1 = group_school["budget"].sum()

mean_budget1 = group_school["budget"].mean()

student_budget1 = mean_budget1 / total_students1 

average_math1 = group_school["math_score"].mean()

average_reading1 = group_school["reading_score"].mean()

passing_math1 = pass_math["school_name"].value_counts()

passing_read1 = pass_reading["school_name"].value_counts()

percent_math1 = (passing_math1 / total_students1) * 100

percent_reading1 = (passing_read1 / total_students1) * 100

school_overall_pass1 = (percent_math1 + percent_reading1) / 2
```

```python
school_summary_df = pd.DataFrame({"School Type" : school_name1,
                                  "Total Students": total_students1, 
                                  "Total School Budget": total_budget1,
                                  "Per Student Budget": student_budget1,
                                  "Average Math Score": average_math1,
                                  "Average Reading Score": average_reading1,
                                  "Percent Passing Math": percent_math1,
                                  "Percent Passing Reading": percent_reading1,
                                  "Percent Overall Passing Rate": school_overall_pass1,
                                  })
```

```python
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)


school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name", as_index=False)
math_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name", as_index=False)
math_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name", as_index=False)
math_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name", as_index=False)
```

```python
math_avg_9th = pd.DataFrame(math_9th["math_score"].mean())
math_avg_10th = pd.DataFrame(math_10th["math_score"].mean())
math_avg_11th = pd.DataFrame(math_11th["math_score"].mean())
math_avg_12th = pd.DataFrame(math_12th["math_score"].mean())
```

```python
math_avg = pd.merge(math_avg_9th, math_avg_10th, on="school_name")
math_avg = pd.merge(math_avg, math_avg_11th, on="school_name")
math_avg = pd.merge(math_avg, math_avg_12th, on="school_name")
math_avg.columns = ["School Name","9th Grade","10th Grade","11th Grade","12th Grade"]

math_avg
```

  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 585, 629, 644, 675]
groups = ["<$585", "$585-629", "$630-644", "$645-675"]
```

```python
school_summary_df["Spending Ranges"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=groups)
```

```python
school_spending = school_summary_df[["Average Math Score", "Average Reading Score", "Percent Passing Math", "Percent Passing Reading",
                                   "Percent Overall Passing Rate", "Spending Ranges"]]
school_spending = school_spending.groupby(["Spending Ranges"]).mean()

school_spending

```
